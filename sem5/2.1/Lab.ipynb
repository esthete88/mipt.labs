```python
import smartlab as sl
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
iu4 = pd.read_csv('iu4.tsv', delimiter='\t')
sl.table(iu4, transpose=True)
```

```python
yerr = sl.abserror(0.5, iu4.iloc[:, 0])
sl.interplot(iu4.iloc[:, 1], iu4.iloc[:, 0], yerr=yerr, xlabel='$V_a,\ V$', ylabel='$I_k,\ \mu A$', name='iu4', save=True)
```

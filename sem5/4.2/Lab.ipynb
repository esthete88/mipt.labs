```python
import pandas as pd
import smartlab as sl
import numpy as np

%matplotlib inline
```

```python
data = pd.read_csv('data.tsv', delimiter='\t')
```

```python
data
```

```python
sl.interplot(data.iloc[:, 3], data.iloc[:, 2], xerr=sl.abserror(40, data.iloc[:, 2]),
             xlabel='$p,\\ KeV/c$', ylabel='$N,\\ s^{-1}$', name='pN', save=True)
```

```python
sl.interplot(data.iloc[:, 3], data.iloc[:, 2] * data.iloc[:, 3], xerr=sl.abserror(40, data.iloc[:, 2]),
             xlabel='$p,\\ KeV/c$', ylabel='$N\\cdot p,\\ KeV/(s*c)$', name='pNp', save=True)
```

```python
sl.linplot(data.iloc[6:-7, 4], data.iloc[6:-7, 5], xerr=sl.abserror(20, data.iloc[6:-7, 4]),
             xlabel='$E,\\ KeV$', ylabel='$FC$', name='FC', save=True)
```

```python
import numpy as np
import smartlab as sl
%pylab inline
```

```python
D = np.array([3.44, 2.38, 1.1, 0.5, 0.47])
d = np.array([37, 56, 111, 220, 250])
```

```python
D1 = 1/D
```

```python
sl.linplot(D1, d, xlabel='$D^{-1},\ mm^{-1}$', ylabel='$d,\ \mu m}$', dpi=150, save=True, name='plot')
```

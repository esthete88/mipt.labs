```python
import numpy as np
import seaborn
import smartlab as sl
import pandas as pd
%pylab inline
```

```python
data = pd.read_csv('data.tsv', delimiter='\t')
data
```

```python
const = 9 * np.pi * sqrt(2*((1.83 * 1e-8 * 0.5) ** 3)/(9.8*898)) * 0.725 * 1e-2
const
```

```python
const * 2/ (20 ** 1.5)/(1.6 * 1e-19)
```

```python
data['$q_1,\ e$'] = const/data.iloc[:,0] * (data.iloc[:,1]+data.iloc[:,2])/(data.iloc[:,1] ** 1.5)/data.iloc[:,2]/1000/(1.6 * 1e-19)
data['$q_2,\ e$'] = const/data.iloc[:,0] * (data.iloc[:,3]+data.iloc[:,4])/(data.iloc[:,3] ** 1.5)/data.iloc[:,4]/1000/(1.6 * 1e-19)
data['$q_3,\ e$'] = const/data.iloc[:,0] * (data.iloc[:,5]+data.iloc[:,6])/(data.iloc[:,5] ** 1.5)/data.iloc[:,6]/1000/(1.6 * 1e-19)
data['$q_4,\ e$'] = const/data.iloc[:,0] * (data.iloc[:,7]+data.iloc[:,8])/(data.iloc[:,7] ** 1.5)/data.iloc[:,8]/1000/(1.6 * 1e-19)
data['$q_5,\ e$'] = const/data.iloc[:,0] * (data.iloc[:,9]+data.iloc[:,10])/(data.iloc[:,9] ** 1.5)/data.iloc[:,10]/1000/(1.6 * 1e-19)
data['$q_6,\ e$'] = const/data.iloc[:,0] * (data.iloc[:,11]+data.iloc[:,12])/(data.iloc[:,11] ** 1.5)/data.iloc[:,12]/1000/(1.6 * 1e-19)
data
```

```python
data['$q,\ e$'] = (data['$q_1,\ e$']+data['$q_2,\ e$']+data['$q_3,\ e$']
                   +data['$q_4,\ e$']+data['$q_5,\ e$']+data['$q_6,\ e$'])/6
data['$\Delta q,\ e$'] = np.sqrt(((data['$q_1,\ e$']-data['$q,\ e$'])**2+(data['$q_2,\ e$']-data['$q,\ e$'])**2+
                                  (data['$q_3,\ e$']-data['$q,\ e$'])**2+(data['$q_4,\ e$']-data['$q,\ e$'])**2+
                                  (data['$q_5,\ e$']-data['$q,\ e$'])**2+(data['$q_6,\ e$']-data['$q,\ e$'])**2)/30+
                                 (data['$q,\ e$']*(2/data.iloc[:,2]+1/data.iloc[:,1]) ** 2))
                          
data
```
